```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DRealme&otracker=nmenu_sub_Electronics_0_Realme'
```

```python
req=requests.get(url)
```

```python
content=BeautifulSoup(req.content,'html.parser')
print(content)
```

# Extracting the parameters of the mobile

```python
name=content.find_all('div',{"class":"_4rR01T"})
price=content.find_all('div',{"class":"_30jeq3 _1_WHN1"})
rating=content.find_all('div',{"class":"_3LWZlK"})
```

```python
print(name[0])
print(price[0])
print(rating[0])
```

```python
n=[]
p=[]
r=[]
m=[]
c=[]
s=[]
```

```python
for i in name:
    n.append(i.text)
for i in price:
    p.append(i.text)
for i in range(len(n)):
    r.append(rating[i].text)
```

```python
 for a in content.find_all('a',class_="_1fQZEK"):
    t=[]
    for i in a.find_all("li",class_='rgWa7D'):
      t.append(i.get_text())
    m.append(t[0])
    s.append(t[1])
    c.append(t[2])
```

```python
data={'NAME':n,'PRICE':p,'RATING':r,"MEMORY":m,"CAMERA":c,"SCREEN":s}
df=pd.DataFrame(data)
```
